```python
from nrem_sc.constants import PROCESSED_DATA_PATH
from nrem_sc.utils import plot_intervals

import pynapple as nap
import matplotlib.pyplot as plt

UNIT_IDS = ['83b', '85b', '116b', '119b']
```

```python
for unit_id in UNIT_IDS:
    fig, axes = plt.subplots(2, 1, figsize=(16, 4), sharex=True)
    sleep_states = nap.load_file(PROCESSED_DATA_PATH / unit_id / "sleep.npz")
    sessions = nap.load_file(PROCESSED_DATA_PATH / unit_id / "sessions_labeled.npz")

    plot_intervals(sleep_states, column='state', ax=axes[0])
    plot_intervals(sessions, column='label', ax=axes[1])
    
    axes[0].set_title(f"{unit_id}")
    axes[1].set_xlabel('')
    plt.tight_layout()
```
