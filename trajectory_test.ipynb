```python
import pandas as pd
import pynapple as nap

from sklearn.preprocessing import StandardScaler
from trajectoryViewer import TrajectoryViewer
from manifold_viewer_3d_fps import TrajectoryViewer3D

fname = '/Users/iii9781/Library/CloudStorage/OneDrive-NorthwesternUniversity/Documents/points3d.csv'
data = pd.read_csv(fname, skiprows=7)
t = data['Time (Seconds)'].to_numpy()
xyz = data[['X.1', 'Y.1', 'Z.1']].to_numpy()
scaler = StandardScaler()
```

```python
def minmax(arr):
    return (arr-arr.min())/(arr.max() - arr.min())

xyz[:,0] = minmax(xyz[:,0])
xyz[:,1] = minmax(xyz[:,1])
xyz[:,2] = minmax(xyz[:,2])
```

```python
tsdf = nap.TsdFrame(t=t, d=xyz, columns=['x','y','z'])
```

```python
import pygfx as gfx

grid = gfx.Grid(
    None,
    gfx.GridMaterial(
        major_step=50,
        thickness_space="screen",
        major_thickness=1,
        infinite=True,
    ),
    orientation="xy",
)
grid.local.y = -10
```

```python
v = TrajectoryViewer3D(background="#ebe2e2")
v.scene.add(grid)
v.add_trajectory(data=tsdf, trail_length=100)
```
